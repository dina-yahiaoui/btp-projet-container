```python
import mysql.connector
import pandas as pd

# Connexion à la base
conn = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="",  # Ton mot de passe si besoin !
    database="chinook"
)

# Exemple pour afficher 10 premières entrées de Artist
df_artist = pd.read_sql("SELECT * FROM Artist LIMIT 10", conn)
display(df_artist)

# Pour Album
df_album = pd.read_sql("SELECT * FROM Album LIMIT 10", conn)
display(df_album)

# Pour Track
df_track = pd.read_sql("SELECT * FROM Track LIMIT 10", conn)
display(df_track)

# Pour Playlist
df_playlist = pd.read_sql("SELECT * FROM Playlist LIMIT 10", conn)
display(df_playlist)

# Pour une association PlaylistTrack (pour voir les liens)
df_playlisttrack = pd.read_sql("SELECT * FROM PlaylistTrack LIMIT 10", conn)
display(df_playlisttrack)

conn.close()

```
